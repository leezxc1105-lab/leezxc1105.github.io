네이버 구글 헤더 정보

```python
import requests

urls = {
    "네이버 메인": "https://www.naver.com",
}

for name, url in urls.items():
    response = requests.get(url)
    print(f"[{name} 헤더 정보]")
    # 주요 헤더 정보만 골라서 출력
    for key, value in response.headers.items():
        if key in ['Server', 'Content-Type', 'Cache-Control', 'P3P']:
            print(f"{key}: {value}")
    print("-" * 30)
```

```python
import requests
import pandas as pd

# 1. 데이터 불러오기
url = "https://jsonplaceholder.typicode.com/posts"
response = requests.get(url)
df = pd.DataFrame(response.json())

# 2. 인덱스가 짝수(0, 2, 4...)인 데이터만 필터링
# index % 2 == 0 은 인덱스를 2로 나눈 나머지가 0인 행을 의미해!
even_df = df[df.index % 2 == 0]

# 3. 엑셀 파일로 저장하기
file_name = "even_posts.xlsx"
even_df.to_excel(file_name, index=False)

print(f"--- '{file_name}' 파일 저장 완료! ---")
print(f"전체 데이터 개수: {len(df)}")
print(f"짝수 인덱스 데이터 개수: {len(even_df)}")

# 상단 5개 데이터 확인
even_df.head()
```

```python
import requests
import pandas as pd
import gradio as gr

def filter_posts(user_id):
    url = "https://jsonplaceholder.typicode.com/posts"
    params = {'userId': user_id}
    response = requests.get(url, params=params)
    df = pd.DataFrame(response.json())

    if df.empty:
        return pd.DataFrame(columns=["결과가 없습니다. 1~10 사이의 숫자를 입력하세요."])
    return df

demo = gr.Interface(
    fn=filter_posts,
    inputs=gr.Number(label="userId 입력 (1~10)"),
    outputs=gr.DataFrame(),
    title="사용자별 게시글 필터링"
)

demo.launch(share=True)
```

```python
import requests
import pandas as pd
import gradio as gr
from google.colab import ai  # 코랩 내장 AI 라이브러리

def translate_first_row(user_id):
    # 1. 7번 과제 데이터 불러오기 (userId로 필터링)
    url = "https://jsonplaceholder.typicode.com/posts"
    params = {'userId': user_id}
    response = requests.get(url, params=params)
    df = pd.DataFrame(response.json())

    if df.empty:
        return "데이터가 없습니다.", "N/A"

    # 2. 처음 1행의 'body' 내용 가져오기
    original_text = df.iloc[0]['body']

    # 3. Agent AI를 사용하여 번역하기 (이미지 속 방식 적용)
    # 텍스트 생성을 통해 번역을 요청하는 프롬프트를 작성해
    prompt = f"Translate the following English text to Korean: {original_text}"
    translated_response = ai.generate_text(prompt)

    return original_text, translated_response

# Gradio UI 설정
demo = gr.Interface(
    fn=translate_first_row,
    inputs=gr.Number(label="userId 입력 (1~10)"),
    outputs=[
        gr.Textbox(label="원본 영문 (첫 행)"),
        gr.Textbox(label="AI 번역 결과")
    ],
    title="Data Translator Agent",
    description="userId를 입력하면 해당 유저의 첫 번째 게시글을 한국어로 번역해줍니다."
)

demo.launch()
```
