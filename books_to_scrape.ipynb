```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
url = "http://books.toscrape.com/"

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

titles = []
books = soup.find_all('h3')

for book in books:
    title = book.find('a')['title']
    titles.append(title)
df = pd.DataFrame(titles[:20], columns=['Book Title'])

print(df)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup('html_content', 'html.parser')

h3_tags = soup.find_all('h3')

title_list = []

for h3 in h3_tags:
    a_tag = h3.find('a')
    if a_tag and a_tag.get('title'):
        title_list.append(a_tag['title'])

print(title_list)
```

```python
all_titles = []

for i in range(1, 6):
    url = f"http://books.toscrape.com/catalogue/page-{i}.html"
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')

    books = soup.find_all('h3')
    for book in books:
        all_titles.append(book.find('a')['title'])

df_all = pd.DataFrame(all_titles[:100], columns=['Book Title'])
print(df_all.head())
```

```python
categories = ['travel_2', 'mystery_3', 'historical-fiction_4', 'sequential-art_5', 'classics_6',
              'philosophy_7', 'romance_8', 'womens-fiction_9', 'fiction_10', 'childrens_11']

book_data = []

for cat in categories:
    url = f"http://books.toscrape.com/catalogue/category/books/{cat}/index.html"
    soup = BeautifulSoup(requests.get(url).text, 'html.parser')

    # 10개씩 추출하여 데이터 저장
    for item in soup.select('.product_pod')[:10]:
        title = item.h3.a['title']
        price = float(re.sub(r'[^0-9.]', '', item.select_one('.price_color').text))

        book_data.append({'Title': title, 'Price': price, 'Category': cat.split('_')[0]})

# 데이터프레임 생성 및 정렬 후 출력
df = pd.DataFrame(book_data)
print("--- 카테고리 통합 가장 비싼 책 Top 10 ---")
print(df.sort_values('Price', ascending=False).head(10))
```
