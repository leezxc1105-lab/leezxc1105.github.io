과제 1. faker를 사용하여 1000명의 데이터를 class를 이용하여 만들어라

```python
from faker import Faker
import pandas as pd
import gradio as gr
```

```python
from faker import Faker
import pandas as pd
import gradio as gr

class User:
  def __init__(self,name,old,adr,number,job):
    self.name = name
    self.old = old
    self.adr = adr
    self.number = number
    self.job =job


fake = Faker('ko_KR')

user_data_list = []
for _ in range(1000):
    new_person = User(
        name=fake.name(),
        old=fake.random_int(min=20, max=80),
        adr=fake.address(),
        number=fake.phone_number(),
        job=fake.job()

    )
    user_data_list.append(vars(new_person))

df = pd.DataFrame(user_data_list)

with gr.Blocks() as demo:
    gr.Markdown("### 📋 Faker를 이용한 1000명 사용자 클래스 데이터")
    gr.Dataframe(df)

demo.launch()
```

과제2) 톰과 제리

```python
class Chaser:
  def __init__(self, name, role):
    self.name = name
    self.role = role
    self.distance = 0
    self.status = "대기 중"

  def following_mode(self, target_name):
        # 상태 변경: 추격 모드로 전환
        self.status = f"{target_name}를 추격 중"

        print(f"[{self.name}]이(가) {target_name}를 쫓아가기 갑니다! (모드: {self.status})")

        # 10번 반복하며 거리(Attribute)를 변화시킴
        for i in range(1, 11):
            self.distance += 30  # 한 번 쫓을 때마다 10m씩 전진
            print(f"{i}번째 추격! 현재 이동 거리: {self.distance}m")
tom = Chaser("Tom", "Chaser")

tom.following_mode("Jerry")

print(f"\n결과: {tom.name}은 총 {tom.distance}m를 이동했습니다.")
```
